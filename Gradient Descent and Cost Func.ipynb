```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model as lm
#Gradient is an efficient algorithm to find the best fit line for the given data set
def gradient_descent(x,y):
    m_curr = b_curr = 0
    iterations = 1000
    n = len(x)
    lr = 0.07
    for i in range(iterations):
        y_pred = m_curr*x + b_curr
        cost = 1/n*(np.sum((y - y_pred)**2))
        md = -(2/n)*(np.sum(x*(y-y_pred)))
        bd = -(2/n)*(np.sum(y-y_pred))
        m_curr = m_curr - lr*md
        b_curr = b_curr - lr*md
        print("m : {},b = {},iter : {},cost = {}".format(m_curr,b_curr,i,cost))        
x = [1,2,3,4,5]
y = [5,7,9,11,13]
x = np.array(x)
y = np.array(y)
gradient_descent(x,y)
```
