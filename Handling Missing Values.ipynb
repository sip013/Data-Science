```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("House_Data.csv")
# print(df.isna())
# print(df.isna().sum())
# print(df.isna().sum().sum())
df2 = df.fillna(value = 0)
# print(df2)
print(df2.isna().sum().sum())
df3 = df.fillna(value = 5)
df
```

```python
#Filling null values with their previous values
df4 = df.fillna(method = 'pad')
df4
```

```python
#Filling the null values with their next value
df5 = df.fillna(method = 'bfill')
df5
```

```python
#filling the values with their previous column values
df6 = df.fillna(method = 'pad',axis=1)
df6
```

```python
#filling the null values with their next column values
df7 = df.fillna(method = 'bfill',axis = 1)
df7
```

```python
#filling the null values column wise according to their respective attributes
col = np.array(df['balcony'])
avg = np.nanmean(col) # this ignores the null values and calculates the avg of the data
df8 = df.fillna({'society':'abcd','balcony':avg})
df8
```

```python
# Dropping the data rows with null values
df9 = df.dropna()
df10 = df.dropna(how = 'any')#Drops those rows which have atleast 1 null value similar to default drop
df11 = df.dropna(how = 'all')#Drops those rows which have all null values
print(df9,df10,df11,sep="\n\n")
```
