```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
num = df.select_dtypes(include = "number") # To include only the numeric columns
print(num.corr())
print(df)
```

```python
sns.heatmap(num.corr()) # to visualise the correlation via heatmap
```

```python
sns.jointplot(x = 'tip',y = 'total_bill',data = df,kind = 'hex')
```

```python
sns.jointplot(x = 'tip',y = 'total_bill',data = df,kind = 'reg') # to form a probability density func.
```

```python
sns.pairplot(df) # this is for multivariable analysis where all the permutations(numeric) are taken and considered by this func.
```

```python
sns.pairplot(df,hue = 'smoker')
```

```python
sns.displot(df['tip']) # Note: distplot is a deprecated func hence we use displot which is same as distplot
sns.displot(df['tip'],kde = True,bins = 10) #helps visualise the pdf of a column [kde - kernel density estimation]
```

```python
sns.countplot(y = 'sex',data = df)
# sns.countplot(x = 'smoker',data = df)
```

```python
sns.countplot(x = 'day',data = df)
```

```python
sns.barplot(x = 'sex', y = 'total_bill',data = df)
```

```python
sns.boxplot(x = df['sex'],y = df['total_bill'])
# sns.boxplot(x = 'sex',y = 'total_bill',data = df)
```

```python
sns.boxplot(x = df['day'],y = df['total_bill'],palette = 'rainbow')
```

```python
sns.boxplot(data = df,orient = 'v',palette = 'rainbow')
```

```python
sns.boxplot(x = df['total_bill'],y = df['day'],hue = df['smoker'])
```

```python
# violin plot : it helps us visualise both the distribution of data in terms of kernel density estimation and the boxplot
sns.violinplot(x = df.total_bill,y = df.day,palette = 'rainbow')
```
