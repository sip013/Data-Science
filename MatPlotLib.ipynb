```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
x = np.arange(1,11)
y = x+1
z = y*x
w = x**2
plt.title("Demo")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.xticks(x)
plt.yticks(z)
plt.plot(y,linestyle = '-',color = 'green',label = "x+1",linewidth = 5,marker = "^",ms = 7,mec = "orange")
#shothand notation : '[color][marker][line]'
plt.plot(z,'r.--',label = "y*x")
plt.plot(w,'b<-',label = "x**2")
plt.legend()
plt.savefig('DemoGraph.jpeg',dpi = 150)
plt.show()
```

```python
#Bar Charts
labels = ['A','B','C']
values = np.random.randint(1,10,size = 3)
plt.title('Bar Graph Demo')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
bar = plt.bar(labels,values)
bar[0].set_hatch('/*')
bar[1].set_hatch('.')
bar[2].set_hatch('O')
plt.figure(figsize = (5,3))
plt.show()
```

```python
gas = pd.read_csv('gas_prices.csv')
# print(type(gas))
plt.figure(figsize = (8,5))
plt.title('Gas Prices over Time(in USD)')
plt.xlabel('Years')
plt.ylabel('USD per Gallon')
for country in gas:
    if(country!='Year'):
        plt.plot(gas['Year'],gas[country],'.-',label = country)
# to project data and visualize it for future..
plt.xticks(gas['Year'][::2].tolist()+[2016])
plt.legend(loc = 1)
plt.savefig('Gas_price_plot.jpeg',dpi = 150)
# Note : using plt.anyfunc() after/before plt.plot() may cause some error if used incorrectly
plt.show()
```
